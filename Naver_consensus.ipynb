```python
import requests
from bs4 import BeautifulSoup
import json 
import pandas as pd

query={
    'flag': '2',
    'cmp_cd': '005930',
    'finGubun': 'IFRSL', #IFRSS
    'frq': '0', #0 연간 #1 분기
    }

url_tmpl = 'https://navercomp.wisereport.co.kr/company/ajax/c1050001_data.aspx?' \
                'flag={flag}&cmp_cd={cmp_cd}&finGubun={finGubun}&frq={frq}'  
url = url_tmpl.format(flag=query['flag'], cmp_cd=query['cmp_cd'], finGubun=query['finGubun'], frq=query['frq'])
headers={'Referer': url}

jo = json.loads(requests.get(url, headers=headers).text)

df = pd.DataFrame(jo['JsonData']).iloc[:, :-1]
df.columns = ['재무연원', '매출액(억)', '매출액(YOY)', '영업이익', '당기순이익', 'EPS', 'BPS', 'PER', 'PBR', 'ROE', 'EV/EBITDA', '주재무제표']
df
```
